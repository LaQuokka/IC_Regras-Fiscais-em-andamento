### Fim do teste

Conclusão: o ```.astype(bool)``` só muda para ```False``` o valor ```0``` especificamente. Qualquer outro valor, mesmo ```NaN``` e ```None```, são tratados como ```1```.

Como precisamos que todos os dados sejam apenas ```0``` ou ```1``` para passar para Bool, vou verificar quantos elementos ```-```, ```NaN```, ```1``` e ```0``` tem em cada coluna que é booleana (as colunas de interesse no momento). Lembrando que todas as colunas devem somar 3922. Se alguma coluna não somar 3922, tem algo errado.

````python
# Criar uma função que conta a quantidade de diferentes valores em cada coluna de interesse

def count_keys(dataframe, coluna):
    vals_e_quants = {} # inicializando um dicionário vazio para guardar os diferentes valores e suas quants
    
    # Iterando pela coluna de interesse
    for row in dataframe[coluna]:
        
        # Se o valor atual já existir no dicionário, ele é incrementado em uma unidade
        if row in vals_e_quants.keys():
            vals_e_quants[row] += 1
        
        # Se o valor atual ainda não tiver sido contabilizado, ele passa a existir no dicionário com o valor 1
        else:
            vals_e_quants[row] = 1
    return vals_e_quants

type_rr = count_keys(df, 'Type_RR')
print(type_rr)
````

````python
quants = []
for coluna in colunas_booleanas:
    dic = count_keys(df, coluna)
    quants.append(dic)
print(quants)
````
